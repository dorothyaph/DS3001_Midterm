```python
import matplotlib.pyplot as plt

models = ["All Features", "Social Media", "Productivity"]
r2_scores = [0.906, 0.000, 0.35]
rmse_scores = [0.583, 1.898, 1.021]

colors = ["#02bf8c", "#fecd6b", "#a7e6ce"]

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# R² Chart
ax[0].bar(models, r2_scores, color=colors)
ax[0].set_title("R² Score (Higher is Better)", fontsize=12, weight="bold")
ax[0].set_ylabel("R²")
ax[0].set_ylim(0, 1)
for i, v in enumerate(r2_scores):
    ax[0].text(i, v + 0.03, f"{v:.2f}", ha="center", fontsize=10)

# RMSE Chart 
ax[1].bar(models, rmse_scores, color=colors)
ax[1].set_title("RMSE (Lower is Better)", fontsize=12, weight="bold")
ax[1].set_ylabel("RMSE")
for i, v in enumerate(rmse_scores):
    ax[1].text(i, v + 0.05, f"{v:.2f}", ha="center", fontsize=10)

plt.suptitle("Regression Model Performance Comparison", fontsize=14, weight="bold")
plt.tight_layout()
plt.show()
```
